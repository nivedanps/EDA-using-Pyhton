```python
#import python libraries

import numpy as np #numpy (slove complex data)
import pandas as pd #pandas(to give explorator form)
import matplotlib.pyplot as plt#matplot for gragh(visualizing)
%matplotlib inline
import seaborn as sns # seaborn
```

```python
#import csv file---csv=coma separated Values
df=pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape #row X columns
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated /blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True) #drop axis=1,inplace=true
```

```python
#check for null vaues
pd.isnull(df).sum()#isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#change data types
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes #.dtypes
```

```python
df.columns#columns
```

```python
#rename column
df.rename(columns={'Martial_Status':'Marriage'})#rename
```

```python
#plotting A bar chart for Gender and it's count
#Gender
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
#if you have to plot a graph showing the total count of values based on categories or group you use these 1
```

```python
#plotting a bar chart fro gender vs total amount(groupby)

sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data =sales_gen)#barplot
```

```python
#plotting A bar chart for Gender and it's count
#Gender
ax=sns.countplot(x='Age',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Gender
ax=sns.countplot(x='Age Group',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python

sales_gen=df.groupby(['Age'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age',y='Amount',data =sales_gen)#barplot
```

```python
#age
ax=sns.countplot(data=df,x='Age Group',hue='Gender') #countplot
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#age Group vs Total amount
sales_gen=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data =sales_gen)
```

```python
#states
#totsa number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)#head(10)
sns.set(rc={'figure.figsize':(20,5)})#seaborn ka stying function
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
#Amount VS STate
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)#head(10)
sns.set(rc={'figure.figsize':(20,5)})#seaborn ka stying function
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
#count plot for Martial status
ax=sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#count plot for Martial status
ax=sns.countplot(x='Marital_Status',data=df)
sns.set( rc={'figure.figsize':(30,15)})#this increases the size of the bar
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data =sales_state)#barplot
```

```python
sales_by_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_by_occupation)
```

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set( rc={'figure.figsize':(30,15)})#this increases the size of the bar
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set( rc={'figure.figsize':(30,15)})#this increases the size of the bar
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_by = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(x='Product_Category', y='Amount', data=sales_by ,color="pink")
```

```python
sales_by = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(x='Product_ID', y='Orders', data=sales_by ,color="red")
```
